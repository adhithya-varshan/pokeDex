```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator,load_img,array_to_img, img_to_array
import tensorflow_datasets as tfds
import os
import tensorflow as tf
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True )

training_Set = train_datagen.flow_from_directory("Poke/train",target_size=(64,64),batch_size=16,class_mode = 'categorical') 
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_Set = test_datagen.flow_from_directory(
        'Poke/val',
        target_size=(64, 64),
        batch_size=32,
        class_mode='categorical')
```

```python
base_model = tf.keras.applications.xception.Xception(weights = "imagenet", include_top = False)
```

```python
avg = tf.keras.layers.GlobalAveragePooling2D()(base_model.output)
output = tf.keras.layers.Dense(151,activation = "softmax")(avg)
model = tf.keras.Model(inputs = base_model.input,outputs = output)
for layer in base_model.layers:
    layer.trainable = False
```

```python
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01,momentum=0.9,name='SGD')
model.compile(loss = 'categorical_crossentropy',optimizer=optimizer,metrics=["accuracy"])
hist = model.fit(training_Set,epochs = 20,validation_data= test_Set)
```

```python
for layer in base_model.layers:
    layer.trainable = True
optimizer = tf.keras.optimizers.SGD(learning_rate=0.001,momentum=0.9,name='SGD')
model.compile(loss = 'categorical_crossentropy',optimizer=optimizer,metrics=["accuracy"])
hist = model.fit(training_Set,epochs = 20,validation_data= test_Set)
```
